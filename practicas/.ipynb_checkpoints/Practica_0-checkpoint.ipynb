## Previo de la Práctica 0

- Definición de Nodo y Topicos en ROS 2.

***-Nodo (Node)***

En ROS 2, un nodo es la unidad fundamental de cómputo. Piensa en un robot complejo como un equipo de especialistas. Cada especialista tiene una única y clara responsabilidad: uno lee los sensores, otro controla los motores de las ruedas, otro planea la ruta, etc. En ROS 2, cada uno de estos "especialistas" sería un nodo: un programa independiente y modular que realiza una tarea específica.

Los nodos son los bloques con los que se construye una aplicación de ROS. Pueden estar todos en la misma computadora o distribuidos en diferentes máquinas, y se comunican entre sí para lograr un objetivo común.


***-Tópico (Topic)***

Un tópico es el canal de comunicación que usan los nodos para intercambiar datos. Si los nodos son los especialistas, los tópicos son como "canales de radio" o "grupos de chat" temáticos.

    Un nodo que quiere compartir información (ej. el nodo de un sensor de cámara) publica mensajes en un tópico específico (ej. /datos_de_camara).

    Cualquier otro nodo que necesite esa información (ej. un nodo de detección de objetos) se suscribe a ese tópico para recibir los mensajes.

Este sistema es muy flexible, ya que un tópico puede tener múltiples publicadores y múltiples suscriptores, permitiendo una comunicación de uno-a-muchos o muchos-a-muchos.

***-  Los Tres Tipos de Interfaces en ROS 2***

En ROS 2, los nodos se comunican a través de tres tipos principales de interfaces: mensajes, servicios y acciones. Cada una está diseñada para un tipo diferente de interacción.

    Mensajes (Messages - .msg): Son para la comunicación unidireccional y continua. Se usan con los tópicos. Un nodo publica mensajes (como datos de un sensor o el estado de un robot) y otros nodos se suscriben para recibirlos. Es ideal para flujos de datos que no necesitan una respuesta directa.

    Servicios (Services - .srv): Son para la comunicación bidireccional de tipo solicitud/respuesta. Un nodo (el cliente) envía una petición a otro nodo (el servidor) y espera hasta recibir una respuesta. Son útiles para tareas rápidas que deben completarse antes de continuar, como consultar el estado de un componente o realizar un cálculo rápido.

    Acciones (Actions - .action): Son para tareas de larga duración que requieren retroalimentación (feedback). A diferencia de los servicios, las acciones son "no bloqueantes", lo que significa que el cliente no tiene que esperar a que la tarea termine. El cliente envía un objetivo (goal), el servidor lo ejecuta y durante el proceso envía retroalimentación constante. Al final, entrega un resultado. Son perfectas para tareas como "navega a un punto" o "mueve el brazo robótico a una posición", ya que se puede monitorear el progreso y cancelarlas si es necesario.



## Desarrollo de la práctica 

-El desarrollo de la práctica consistió en la creación de un paquete de ROS 2 llamado nodes_topics_py y la programación de tres nodos en Python para simular el control de velocidad de un motor. A continuación se presenta el código final de cada componente del sistema.

***1. Nodo Publicador (rpm_publisher.py)***

Este nodo simula un motor, publicando una velocidad variable en RPM con una función senoidal en el tópico /rpm_topic.

```python
#!/usr/bin/env python3
import rclpy
from rclpy.node import Node
from std_msgs.msg import Float64
import math

class RpmPublisher(Node):
  def __init__(self):
    super().__init__("rpm_publisher_node")
    self.amplitude_ = 60.0
    self.avg_rpm_ = 120.0
    self.frequency_ = 0.1
    self.start_time_ = self.get_clock().now()
    self.publisher_ = self.create_publisher(Float64, "rpm_topic", 10)
    self.timer_ = self.create_timer(0.1, self.publish_rpm)
    self.get_logger().info("Nodo publicador de RPM (senoidal) activo.")

  def publish_rpm(self):
    elapsed_time = (self.get_clock().now() - self.start_time_).nanoseconds / 1e9
    rpm_value = self.amplitude_ * math.sin(2 * math.pi * self.frequency_ * elapsed_time) + self.avg_rpm_
    msg = Float64()
    msg.data = rpm_value
    self.publisher_.publish(msg)

def main(args=None):
  rclpy.init(args=args)
  node = RpmPublisher()
  rclpy.spin(node)
  rclpy.shutdown()

if __name__ == '__main__':
  main()
```


***2. Nodo Suscriptor-Publicador (rpm_converter_node.py)***

Este nodo se suscribe al tópico /rpm_topic, convierte el valor de RPM a radianes por segundo (rad/s), y publica el resultado en el tópico /rad_s_topic.

```python
#!/usr/bin/env python3
import rclpy
from rclpy.node import Node
from std_msgs.msg import Float64
import math

class RpmConverter(Node):
  def __init__(self):
    super().__init__("rpm_converter_node")
    self.subscription_ = self.create_subscription(Float64, "rpm_topic", self.rpm_callback, 10)
    self.publisher_ = self.create_publisher(Float64, "rad_s_topic", 10)
    self.get_logger().info("Nodo conversor de RPM a rad/s activo.")

  def rpm_callback(self, msg):
    rpm_value = msg.data
    rad_s_value = rpm_value * (2 * math.pi) / 60.0
    self.get_logger().info(f'Recibido: {rpm_value:.2f} RPM -> Calculado: {rad_s_value:.2f} rad/s')
    new_msg = Float64()
    new_msg.data = rad_s_value
    self.publisher_.publish(new_msg)

def main(args=None):
  rclpy.init(args=args)
  node = RpmConverter()
  rclpy.spin(node)
  rclpy.shutdown()

if __name__ == "__main__":
  main()
```

***3. Nodo Suscriptor (speed_display_node.py) SOLO PARA VERIFICAR SU FUNCIONAMIENTO CORRECTO SE CREO ESTE TERCER NODO***

Este nodo se suscribe al tópico /rad_s_topic para recibir el resultado final y mostrarlo en la consola, permitiendo verificar el correcto funcionamiento del sistema.

```python
#!/usr/bin/env python3
import rclpy
from rclpy.node import Node
from std_msgs.msg import Float64

class SpeedDisplay(Node):
  def __init__(self):
    super().__init__("speed_display_node")
    self.subscription_ = self.create_subscription(Float64, "rad_s_topic", self.display_callback, 10)
    self.get_logger().info("Nodo display de velocidad en rad/s activo.")

  def display_callback(self, msg):
    self.get_logger().info(f'Velocidad final del motor: {msg.data:.2f} rad/s')

def main(args=None):
  rclpy.init(args=args)
  node = SpeedDisplay()
  rclpy.spin(node)
  rclpy.shutdown()

if __name__ == "__main__":
  main()
```


***4. Modificación de los Archivos de Configuración del Paquete***

Para que el sistema de ROS 2 pueda reconocer, compilar y ejecutar nuestros nodos de Python, fue crucial configurar correctamente los dos archivos principales del paquete: package.xml y setup.py.

***package.xml (El Manifiesto del Paquete)***

Este archivo es el "manifiesto" del paquete. Define sus propiedades, como el nombre, la versión, el autor y, lo más importante, sus dependencias.

Para este proyecto, nos aseguramos de que el archivo declarara explícitamente que nuestro paquete depende de rclpy (la librería cliente de ROS 2 para Python) y std_msgs (que contiene los tipos de mensajes estándar como Float64 que usamos para enviar los datos de velocidad).

```python
<?xml version="1.0"?>
<?xml-model href="http://download.ros.org/schema/package_format3.xsd" schematypens="http://www.w3.org/2001/XMLSchema"?>
<package format="3">
  <name>nodes_topics_py</name>
  <version>0.0.0</version>
  <description>Paquete para la Práctica 0: Nodos y Tópicos en Python</description>
  <maintainer email="erik.pena@ingenieria.unam.edu">robousr</maintainer>
  <license>MIT</license>

  <depend>rclpy</depend>
  <depend>std_msgs</depend>

  <test_depend>ament_copyright</test_depend>
  <test_depend>ament_flake8</test_depend>
  <test_depend>ament_pep257</test_depend>
  <test_depend>python3-pytest</test_depend>

  <export>
    <build_type>ament_python</build_type>
  </export>
</package>
```



***setup.py (El Instalador)***

Este archivo es el script de instalación para un paquete de Python. Le indica a la herramienta de construcción de ROS 2, colcon, cómo instalar nuestros nodos para que podamos ejecutarlos desde la terminal con el comando ros2 run.

La modificación más importante se realizó en la sección entry_points. Aquí, registramos cada uno de nuestros tres scripts de Python como un "punto de entrada", asignándole un nombre de ejecutable. Esto es lo que nos permite llamar a nuestros nodos por sus nombres (rpm_publisher, rpm_converter, speed_display) desde la consola.

```python
from setuptools import find_packages, setup

package_name = 'nodes_topics_py'

setup(
    name=package_name,
    version='0.0.0',
    packages=find_packages(exclude=['test']),
    data_files=[
        ('share/ament_index/resource_index/packages',
            ['resource/' + package_name]),
        ('share/' + package_name, ['package.xml']),
    ],
    install_requires=['setuptools'],
    zip_safe=True,
    maintainer='robousr',
    maintainer_email='erik.pena@ingenieria.unam.edu',
    description='Paquete para la Práctica 0: Nodos y Tópicos en Python',
    license='MIT',
    tests_require=['pytest'],
    entry_points={
        'console_scripts': [
            # Se define el nombre del ejecutable y se le asocia el archivo y la función main()
            'rpm_publisher = nodes_topics_py.rpm_publisher:main',
            'rpm_converter = nodes_topics_py.rpm_converter_node:main',
            'speed_display = nodes_topics_py.speed_display_node:main'
        ],
    },
)
```

***Estas configuraciones son esenciales para integrar un paquete de Python en el ecosistema de ROS 2 y asegurar que funcione correctamente.***
